```python
from random import randint
import matplotlib.pyplot as plt
import numpy as np
```

```python
sets = [1, 2, 3, 4, 5, 6]
```

```python
def tigerMachine(sb, tiger):
    sb = sb
    tiger = tiger
    sbInit = sb
    tigerInit = tiger
    num = 0
    win = 0
    while sb >= 2.99 and tiger >= 0:
        count = 0
        tiger += 2.99
        sb -= 2.99
        die1 = randint(1,6)
        die2 = randint(1,6)
        if die1 == 1 and die2 in sets:
            sb += 0.88
            if tiger < 0.88:
                break
            tiger -= 0.88
        elif die1 == 2 and die2 in sets[:-1]:
            sb += 1.88
            if tiger < 1.88:
                break
            tiger -= 1.88
        elif die1 == 3 and die2 in sets[:-2]:
            sb += 3.88
            if tiger < 3.88:
                break
            tiger -= 3.88
        elif die1 == 4 and die2 in sets[:-3]:
            sb += 8.88
            if tiger < 8.88:
                break
            tiger -= 8.88
        elif die1 == 5 and die2 in sets[:-4]:
            sb += 12.88
            if tiger < 12.88:
                break
            tiger -= 12.88
        elif die1 == 6 and die2 in sets[:-5]:
            sb += 28.88
            if tiger < 28.88:
                break
            tiger -= 28.88
        if tiger >= tigerInit:
            win += 1
        num += 1
        #if sb < tiger:
        #    count = 1
       # print(num, round(sb, 1), round(tiger, 1), count)
    return [num, win, round(sb, 2), round(tiger, 2)]  
```

```python
testSets = []
for i in range(100):
    testSets.append(tigerMachine(100, 100)[0])
min(testSets)
plt.plot(testSets)
plt.show()
```

```python
# 游戏次数对盈利的影响
def timesForTigerMachine(times, sb, tiger):
    win = 0
    times = times
    num = 0
    for i in range(times):
        results = tigerMachine(sb, tiger)
        if results[2] < results[3]:
            win += 1
        num += results[0]
    return win, times, win/times, num/times
```

```python
# 本金对盈利概率的影响
def tigerMachineP(times, sb, tiger):
    win = 0
    times = times
    num = 0
    for i in range(times):
        results = tigerMachine(sb, tiger)
        if results[2] < results[3]:
            win += 1
        num += results[0]
    return win, times, win/times, num/times
```

```python
def timesForTigerMachine(sb, tiger, times):
    sb = sb
    tiger = tiger
    sbInit = sb
    tigerInit = tiger
    num = 0
    win = 0
    times = times
    while sb >= 2.99 and times > 0:
        count = 0
        tiger += 2.99
        sb -= 2.99
        times -= 1
        die1 = randint(1,6)
        die2 = randint(1,6)
        if die1 == 1 and die2 in sets:
            sb += 0.88
            if tiger < 0.88:
                break
            tiger -= 0.88
        elif die1 == 2 and die2 in sets[:-1]:
            sb += 1.88
            if tiger < 1.88:
                break
            tiger -= 1.88
        elif die1 == 3 and die2 in sets[:-2]:
            sb += 3.88
            if tiger < 3.88:
                break
            tiger -= 3.88
        elif die1 == 4 and die2 in sets[:-3]:
            sb += 8.88
            if tiger < 8.88:
                break
            tiger -= 8.88
        elif die1 == 5 and die2 in sets[:-4]:
            sb += 12.88
            if tiger < 12.88:
                break
            tiger -= 12.88
        elif die1 == 6 and die2 in sets[:-5]:
            sb += 28.88
            if tiger < 28.88:
                break
            tiger -= 28.88
        if tiger >= tigerInit:
            win += 1
        num += 1
        #if sb < tiger:
        #    count = 1
        #print(num, round(sb, 1), round(tiger, 1), count)
    return [num, win, round(sb, 2), round(tiger, 2)]  
```
