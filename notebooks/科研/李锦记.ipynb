牛肉中的蛋白质、脂肪、核糖及其降解产物在受热过程中发生反应，产生的香气物质，是肉提取物中存在的化合物经加热分解的产物，主要是美拉德反应和史氏递降分解作用引起的分解产物。一是增加风味，二是矫正或调整原料肉的不快味道[12]。传统牛肉粒最终基本保持原料牛肉的纤维结构，纤维结构经干制后，形成了难以咀嚼的口感[13]。原料肉筋膜处理不完全，经干制后会形成更难咀嚼的组织，影响产品的口感。这些使传统牛肉粒食品成为年轻人的美食，渴望尝试的老年和小孩只能望而却步，影响产品的消费群体和效益。本产品粗粮牛肉粒采用绞碎工艺与粗粮结合，打破了牛肉纤维组织，产品质地酥软，易咀嚼，并且添加李锦记蜜汁烤肉酱大大提高了产品的口感，扩大了产品的接受人群，且不影响传统肉干对咀嚼性的要求。

财富的保值增值是一个人们永远乐此不疲的话题，但我发现在和很多朋友讨论投资理财时，大家对风险发生的概率缺乏一个真正理性的认识。某次一位朋友咨询我关于股票投资的问题，我告诫他股票投资有风险，股市盈利的概率可能不足 10%。他用真诚的目光看着我说：「我当然知道投资股市有很大的失败概率，所以我首次投资希望可以走得稳妥一点，总不能一上来就亏损吧，这样你要我如何跟老婆交代？」对这位抱着这种前后自相矛盾的逻辑的朋友，我真的很想给他推荐《超预测》这本书。

我这位友人想赢怕输的心理恰恰映射了我们大多数人的某种概率认知谬误。人们在判断某事时很容易受心理学的锚定效应影响，以初始的几次状态来判断事物的优劣，比如在职场上如果一位新人上来就出色地完成了几个销售任务，人们就会给予他极高的评价，即便之后他未能保持如此惊艳的水准，人们也更倾向于将这归咎于他「表现失常」。相反，即便是一位资深的业务人员，如果他搞砸了自己在某地的首秀，便很容易被大家冠以言过其实等评价。须知即便我说此事十拿九稳，即有 90% 的成功概率，它仍然有 10% 的失败可能性，而人们往往忽略这个因素的存在。再回看我那位朋友的理财观念，即便他能找到巴菲特为他买卖股票，他也很大概率会在首次投资中遭遇亏损，那我们能因此断定巴菲特也是欺世盗名之徒吗？当然不会，因为巴菲特的厉害之处在于可以在 40 年的跨度之内维持稳定的收益。

因此在面对不确定事物的时候，与其纠结于每一次的得失，不如思考如何提升系统的成功概率，并学会在多次样本的观测中来评估事物的靠谱程度，切忌用初印象就武断地给问题下结论，这样得出的结论往往会和事实南辕北辙。

当然这本书对我们的启迪远不止于此，我认为它给我带来的最大益处就是全面重塑了我对事物发展的判断方式和对概率预测逻辑的认知，告诉读者如何告别拍脑袋式的决策方式，让你的分析和预测真的靠谱起来。如果你希望提升自己的判断力和逻辑分析能力，那这本书会给你想要的答案。

接下来简单地为大家介绍下这本书的作者。本书作者菲利普•泰洛克是著名心理学家，全球最受关注的社会学家之一，宾夕法尼亚大学教授，同时还在沃顿商学院心理学和政治学部门任职。某些专业预测人士确实具有远见卓识，泰洛克一直尝试着找出原因并探究原因背后的机制。他和妻子芭芭拉·梅勒斯是「精准预测项目」的共同负责人，这是一项已开展多年的预测研究项目。当然也正是这个项目让泰洛克感悟到了许多真谛并完成了《超预测》这本书。

上面介绍完了这本书的基本情况和作者。接下来，我就为你详细说一下这本书的内容。这本书带给读者最大的启发主要有以下三点：

1. 没有什么是确定的，唯一确定的只有不确定性； 

2. 问题拆解式的系统预测模型；

3. 看待问题时一定要先从外到内进行分析，不要被细节束缚。

我先来给大家介绍第一个要点：没有什么是确定的，唯一确定的只有不确定性。

作者在这本书的开篇提到了一个普遍存在却又被忽视的现象，即不同的人对同一个概率的表述总是存在误解。例如你拜托我一件事，我告诉你这件事我很有把握把它搞定，那这个很有把握意味着什么呢，它的实际概率是多少呢？可能以我的标准而言，很有把握意味着 70% 的成功概率，但在你的标准里很有把握意味着 100% 的成功概率。最后如果这件事情失败了，我认为也完全可以接受，毕竟 70% 的成功概率同时表明还有 30% 的概率失败，但是在你看来这样的结果就是难以接受的。

果胶酶制剂和相关酶的常见用途包括组织浸渍，组织液化，增强的回收或提取（汁液或油），澄清和促进剥离（尤其是柑橘类水果）。商业“果胶酶”制剂通常是几种类型的果胶，纤维素和半纤维素降解酶的粗混合物，并且“果胶酶”应用中的进化将针对特定目的和产品的更具体的酶部署。在大多数情况下，酶处理在20-30℃下进行数小时以进行温和处理，或在40-50℃下进行1-2小时进行更严格的处理（浸渍，液化）在果汁的pH值或提取。组织处理开始于组织的初始破碎（研磨）或粗磨（图6.21 [48,77,142]）。粗组织制备将制备专门水解和部署中间薄片（果胶物质）的制剂，因为这将最容易地释放具有完整细胞壁的单个细胞（图6.21a）。因此，内多聚半乳糖醛酸酶（特别是）或内切蛋白裂解酶是最合适的，并且所得的细胞悬浮液可用于果汁或花蜜，婴儿食品或其他产品的成分。

其他加工流源于组织的细磨或破碎（通常通过锤磨机），其中液体的最大排出，有时是稀释物质，是期望的结果。来自梨果实和浆果的“提取物”通常需要添加酶以将粘稠的果肉转化为粉碎的果实的半凝胶糊状物（由果胶的部分溶解和固体的高保水能力引起），以在随后的压榨过程中最大化果汁提取（图6.21b）。能够解聚和降解高度甲氧基化果胶的果胶降解酶是最合适的，并且特别包括内切半乳糖醛酸酶和果胶甲基酯酶，同时也可以使用内切蛋白裂解酶。以这种方式制备的果汁可以是澄清的或混浊的，这取决于所使用的特定组织和酶的组合。

水果或植物材料的液化用于将整个组织块转化为液体产品，并且这种产品不需要过滤或压制（图6.21c）。这是通过果胶酶（多聚半乳糖醛酸酶，果胶甲基酯酶和果胶裂解酶），纤维素酶（外切和内切β-葡聚糖酶）和半纤维素酶（作用于木聚糖，甘露聚糖，半乳聚糖和阿拉伯聚糖）的强有力组合来实现的。一旦大部分中间薄片和细胞壁材料“溶解”（多达80％），细胞很容易通过渗透压或剪切破裂以释放液体内容物。液化用于将许多稀烂的热带水果（芒果，番石榴，香蕉），橄榄和储存的苹果转化为果汁或含油提取物。取决于所使用的组织和酶，这些果汁可以变浑浊或澄清。

外源果胶和细胞壁降解酶的最后一个主要应用是用于果汁或提取物的分离（图6.21d）。这需要启动破坏果汁或提取物中任何“云”的事件。在一些果汁（例如橙色）中可能需要云，但不适用于那些优选透明果汁的果汁如苹果和葡萄。云由胶体颗粒赋予，所述胶体颗粒由涂有果胶的蛋白质（在汁液pH下带正电荷）组成（半乳糖醛酸残基部分解离并带负电）。果胶解聚酶溶解和破坏果胶层，使蛋白质与其他颗粒的果胶层静电相互作用，导致聚集，发生絮凝，并提供易于克隆。该过程的最佳pH值估计为~3.6，并且通常由果胶甲基酯酶促进裂解，特别是与内多聚半乳糖醛酸酶组合，或者单独的果胶裂解酶用于高度甲氧基化的果胶（苹果）。果胶甲基酯酶作用的作用是产生Ca 2+诱导的颗粒交联的位点，导致易于沉降的聚集体。在某些情况下，蛤蜊果汁可能会发生回复雾霾;阿拉伯糖包含约90％的参与该雾度的多糖物质。如果在压榨和/或煅烧处理中包括内阿拉伯聚糖酶，则雾度最小化。

在柑橘（橙色）果汁中，称为“纸浆洗涤”的过程需要果胶酶，以在将残余纸浆加回到从组织表达的初始汁液之前降低残余纸浆的水提取物的粘度。柑橘还可以通过对表面进行刻划和在20-40℃下真空浸泡果胶酶约1小时进行酶去皮，因为这样可以消化白色海绵状反照率，从而在去除皮肤后容易分割“去皮”果实。 （FL avedo）。有关用于特定水果的过程和酶以及由其制备的产品的具体考虑因素存在并在别处详细描述[142]。
