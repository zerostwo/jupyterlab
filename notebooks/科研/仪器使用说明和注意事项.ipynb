# 万用电炉

## 工作原理

电流通过电阻丝产热。

## 用途说明

用于加热。

## 使用说明

1. 检查万用电炉功率调节旋钮是否完全旋转到左面，若无，旋转功率调节旋钮到最左面；
2. 接通电源；
3. 根据需要调节功率旋转旋钮。旋钮转到最右位置功率最大，反之最小；
4. 使用技术后记得先将旋钮旋到最左端，再关闭电源。

## 注意事项

1. 使用过程中注意安全，防止被烫伤；
2. 使用过程中轻拿轻放；
3. 由于我们实验室的万用电炉的电阻丝是暴露在外面的，所以当炉盘内电热丝处于发红状态时，在其周围环境空气中应无易燃性，腐独性气体或导电尘埃存在。

# 隔膜真空泵（便携式）

## 工作原理

电机带动泵内的旋片，压缩泵内空气，将外界空气吸入腔体。

## 用途说明

用于真空或溶剂过滤等。

## 使用说明

1. 检查真空泵电源开关是否关闭，若未关闭，关闭开关；
2. 接通电源；
3. 打开开关即可使用；
4. 使用结束后先关闭开关，再拔掉电源插头。

## 注意事项

1. 使用过程中注意安全；
2. 抽滤时注意检查输气管是否牢固，保证装置密封性。

# 循环水式多用真空泵

## 工作原理

电机带动泵内叶片，改变腔内体积，吸入外界空气。

## 用途说明

用于为蒸发、蒸馏、结晶、干燥、升华、过滤减压、脱气等过程提供真空条件。

## 使用说明

1. 使用前检查仪器水箱内是否有水。若无水，添加水至水位线；
2. 检查开关是否处于关闭状态；
3. 接通电源，打开开关即可启用抽真空作业；
4. 使用结束后先关闭开关，再拔掉电源插头。

## 注意事项

1. 使用过程中注意安全；
2. 若吸气效果不佳，一方面可以检查水箱内水质情况，水被污染会影响抽真空效果，另一方面检查输气管是否紧密链接；
3. 需要定期更换水箱内的水。


# pH计

## 工作原理

通过测定待测液中的氢离子浓度来转化为待测溶液的pH或电势。

## 用途说明

用于测定待测液pH值或电势。

## 使用说明

1. 插上电源，按压启动按钮；
2. 每次测定样品前，使用超纯水清洗玻璃电极，然后用擦镜纸轻轻吸取电击上的水，随后方可进行测量；
3. 使用结束后，先关闭电源按钮，再拔掉电源插头。

## 注意事项

1. 若pH计长期未用，使用前务必使用标准液校准；
2. 测量过酸（pH<2）或过碱（pH>12）样品后需要重新校准；
3. 使用擦镜纸处理玻璃泡的时候一定要轻轻的，保护好玻璃泡。
